```python
loaded_SKT = pickle.load(open('../Simultaneous_CompatSKT.p', 'rb'))
loaded_DCS = pickle.load(open('../Simultaneous_DCS.p', 'rb'))
```

```python
def mergeAll(paths, csvfn):
    accuracies = {}
    
    for path in paths:
        fNames = os.listdir(path)
        fNames = list(filter(len, fNames))
        i = 0
        for fName in fNames:
            act = pickle.load(open(path + fName, 'rb'))
            
            
            for f, pair in act.items():
                i += 1
                if i % 400 == 0:
                    print('checkpoint', i)
                skt = loaded_SKT[f]
                (chunkDict, lemmaList, wordList, revMap2Chunk, qu, cngList, verbs, tuplesMain, qc_pairs) = SentencePreprocess(skt)
                accuracies[f] = [f, pair[0], pair[1], pair[2], len(lemmaList), len(set(lemmaList))]
    with open(csvfn, 'w') as ff:
        csvF = csv.writer(ff)
        csvF.writerow(['file', 'matched', 'inDCS', 'totalPredicts', '#lemma_nodes', '#unique_lemma_nodes'])
        for line in accuracies.values():
            csvF.writerow(line)
    print(len(accuracies))
```
