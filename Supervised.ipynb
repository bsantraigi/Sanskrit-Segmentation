```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import KFold
from IPython.display import display
```

```python
df_bin = pd.read_csv('extras/metPerfFH_bin.csv')
df_rank = pd.read_csv('extras/metPerfFH.csv')
df_bin = df_bin[df_bin['w2w'] + df_bin['t2t'] + df_bin['w2w_samecng'] > 0]
df_rank['w2w'] = 1/(1 + df_rank['w2w'])
df_rank['t2t'] = 1/(1 + df_rank['t2t'])
df_rank['w2w_samecng'] = 1/(1 + df_rank['w2w_samecng'])
```

```python
df_bin
```

```python
df_rank
```

```python
X = np.mat(df_bin[['w2w', 't2t', 'w2w_samecng']])
Y = np.array(df_bin['InDCS'])
kf = KFold(Y.shape[0])

lr_model = LogisticRegression(solver='lbfgs', fit_intercept=False)
for trainIndex, testIndex in kf:
    lr_model.fit(X[trainIndex, :], Y[trainIndex])
    y2 = lr_model.predict(X[testIndex, :])
    print(np.sum(y2 == Y[testIndex])/len(testIndex))
    
display(lr_model.coef_)
display(lr_model.intercept_)
display(lr_model.n_iter_)
```
